```python
def function1(df):
    df["created_at"] = pd.to_datetime(df["created_at"])
    freq = df.groupby(df["created_at"].dt.date)["text"].count()
    fig, ax = plt.subplots(figsize=(10, 5))
    ax.plot(freq.index, freq.values, marker="o")
    ax.set_xlabel("Date")
    ax.set_ylabel("Fr√©quence des Tweets")
    ax.set_title("Fr√©quence des Tweets par jour")
    ax.grid(True)
    
    for x, y in zip(freq.index, freq.values):
        label = "{:.0f}".format(y)  
        ax.annotate(label, (x, y), textcoords="offset points", xytext=(0,10), ha='center')
    return plt
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from nltk.corpus import stopwords
import nltk

# T√©l√©charger la liste de stop words en fran√ßais
nltk.download('stopwords')

# R√©cup√©rer les textes de la colonne "text2" du dataframe df
textes2 = df['text2'].tolist()

# Fusionner tous les textes en une seule cha√Æne
texte_concatene = ' '.join(textes2)

# Enlever les stop words
stop_words = set(stopwords.words('french'))
texte_sans_stop_words = ' '.join([mot for mot in texte_concatene.split() if mot.lower() not in stop_words])

# Cr√©er l'objet WordCloud et le visualiser
wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='magma', stopwords=None, max_words=200).generate(texte_sans_stop_words)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
